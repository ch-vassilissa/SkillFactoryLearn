```python
from random import randint


def motion_comp():  # функция определения координат хода компа
    xc, xy = None, None
    if step == 2 and not (A[1][1] == 'X'): # займем центр
        xc, xy = 2, 2
    else:
        while True: #перебор пока не найдем пустую клетку
            xc, xy = randint(1, 3), randint(1, 3)
            if A[xc - 1][xy - 1] == '-':
                break
    print('Ход компьютера: ',xc,',',xy)
    return xc, xy


def check_win(play):  # функция поиска линии для игрока = play
    for i in range(0, N):  # пройтись по строкам массива
        if all(elem == play for elem in A[i]):
            end = True
            break
        else:
            end = False

    if not(end):
        for j in range(0, N):  # пройтись по столбцам массива
            for i in range(0, N):
                if not (A[i][j] == play):
                    end = False
                    break
                else:
                    end = True
            if end:
                break

    if not (end):
        for i in range(0, N):  # пройтись по диагонали массива
           if not (A[i][i] == play):
               end = False
               break
           else:
               end = True

    if not (end):
        for i in range(0, N):  # пройтись по диагонали массива
           if not (A[i][N - 1 - i] == play):
               end = False
               break
           else:
               end = True
    return end


def print_mas():  # печать массива
    print ('      1    2    3')
    for i in range(N):
        print(i + 1, '|', A[i])

# создадим матрицу
N = 3  # размерность массива
A = []
step = 1
finish = False

for i in range(0, N):
    A.append([])
for i in range(0, N):
    for j in range(0, N):
        A[i].append(j)
        A[i][j] = '-'
print_mas()

while finish or step < 10:
    x_play, y_play = map(int, input("Введите координаты поля 3*3 (формат x y): ").split())
    if not (1 <= x_play <= 3) or not (1 <= y_play <= 3):
        x_play, y_play = map(int, input("Ошибка ввода. Введите координаты поля 3*3 (формат x y): ").split())
    A[x_play - 1][y_play - 1] = 'X'
    step += 1
    finish = check_win('X')
    print_mas()
    if finish:
        print('Ура - Вы выиграли')
        break

    x_comp, y_comp = motion_comp()
    A[x_comp - 1][y_comp - 1] = 'О'
    step += 1
    finish = check_win('О')
    print_mas()
```
